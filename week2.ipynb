```python
#function for adding a new column in the dataframe named 'PV' which is the multiplication of Closing Price and Volume
# here volume = Total Traded Quantity.
def compute_PV():
    for y in range(0,len(df1)):
        df1['PV'] = (df1['Close Price'] * df1['Total Traded Quantity'])
```

```python
#using .groupby() function for grouping the data using two parameters namely 'Year' and 'Month'.
# .groupby() is function that works on split-apply-combine theory. Here we now have groups of different months of different year
p =df1.groupby(['Year','Month'])
```

```python
#.first() function shows the first value of each month od each year (in short each group!!)
p.first()
```

```python
# .groupby() function has a 'feature' syntax in which we can choose any feature and can do computation on it.
# so we chose here PV column to do the computation of sum. After that, we applied .to_frame() function which converts this into
# a dataframe. And finally we used .reset_index to reset the indexing of the dataframe. When you print the value of 'm', you'll
# see a dataframe having index till 24 which shows that the total number of groups are 25(0-24).
# We used this approch because on iterating directly into groups, the loop was creating a tuple which was much more complex to 
# work with here.
m = p['PV'].sum().to_frame().reset_index()
```

```python
# we did the same thing here too like above. The only difference is that here the feature is 'Total Traded Quantiy' which is 
# also called as 'Volume' in stock market language.
n = p['Total Traded Quantity'].sum().to_frame().reset_index()
```

```python
# Here we have created a list of the values in 'm' by iterating through 'm' using its index values.
lis1 = []
for ind in m.index: 
    lis1.append(m['PV'][ind])
print(lis1)   
```

```python
# similarly we did it for the other feature.
lis2 = []
for ind in n.index: 
    lis2.append(n['Total Traded Quantity'][ind])
print(lis2)   
```

```python
#finally we divided the two lists as per the formula for VWAP since the length of the two lists were same!
res = [i / j for i, j in zip(lis1, lis2)] 
print(res)
```

```python
# converted the given list into an array.
arr = np.array(res)  #VWAP Array generated
print(arr) 
```

```python
# now here we are defining a function for profit/loss percentage change over the last N days
def compute_PCT(m):
    nex = 0
    prev = 0
    df1_copy = df1.copy()
    nex = df1_copy.iloc[493]['Close Price']
#     print(nex)
    a = df1_copy.at[493,'Date'] -  timedelta(days = (m-1))
#     print(a)
    b = df1_copy[df1_copy['Date']==a].index
#     print(b)
    d = b[0]
    prev = df1_copy.at[d,'Close Price']
#     print(prev)
    res = ((nex/prev)-1)*100
    print(res)
    
```

```python
# now we have to calculate the profit/los percentage in closing prices for each day using an inbuilt function called "pct_change()"
df1['Day_Perc_Change']= df1['Close Price'].pct_change()
```

```python
# the value generated would be multiplied by 100 to get the actual percentage value.
df1['Day_Perc_Change'] = df1['Day_Perc_Change'] * 100
```

```python
# Here we can see that the first value is NaN. The reason behind it is that the profit/loss percenatge for a given day is 
# calculated using the price of the previous day. And since the first column is first so it doesn't have any previous value
# of the price. Hence it evaluates to be a NaN.
df1
```

```python
#replacing the first value that is NaN value with 0.
df1['Day_Perc_Change'].fillna(0, inplace=True)
```

```python
# checking the replacement of NaN values.
df1.head()
```

```python
# Here to generate a different column of "Trend", we needed to have multiple conditions to generate multiple types of trends 
# given. Hence we defined a new function and applied that function on the "Day_Perc_Change" column and saved the corresponding
# values in a new column named "Trend".
def trend(x):
    if -0.5<x<0.5:
        return "Slight or No Change"
    elif 0.5<x<1:
        return "Slight Positive"
    elif -1<x<-0.5:
        return "Slight Negative"
    elif 1<x<3:
        return "Positive"
    elif -3<x<-1:
        return "Negative"
    elif 3<x<7:
        return "Among Top Gainers"
    elif -7<x<-3:
        return "Among Top Losers"
    elif x>7:
        return "Bull run"
    elif x<-7:
        return "Bear drop"
# print(trend(-1.86))
df1['Trend'] = df1['Day_Perc_Change'].apply(trend)
```
