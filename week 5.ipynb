```python
#importing important libraries
import numpy as np
import pandas as pd
```

```python
#reading csv files
df = pd.read_csv('week3.csv')
df[5:20]
```

```python
#dropping all rows with NaN values
df = df.dropna()
df
```

```python
#calculating the 'Middle Bollinger band'
df['Mid'] = (df['Upper']+df['Lower'])/2
df
```

```python
# making the 'Call' column by defining a function.
def call(df):
    if df['Close Price'] < df['Lower']:
        return "Buy"
    if df['Close Price'] > df['Lower'] and df['Close Price'] < df['Mid']:
        return "Hold Buy/ Liquidate Short"
    if df['Close Price'] > df['Mid'] and df['Close Price'] < df['Upper']:
        return "Hold Short/ Liquidate Buy"
    if df['Close Price'] >df['Upper']:
        return "Short"
df = df.assign(Call = df.apply(call, axis=1))
df
```

```python
# command for displaying all the rows of a dataset.
pd.set_option("display.max.rows", None)
```

```python
#saving the training data in a different dataset
train_X = df[['Upper','Lower','Mid','Close Price']]
```

```python
#analysing training dataframe.
train_X
```

```python
#now since our output is in categorical data, hence we need to encode it. We are using here label encoder for the same. 
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
transformed_label = le.fit_transform(df[['Call']])
```

```python
transformed_label
```

```python
#now since this is 'transformed_label' is a 1D array, we are converting it into 2D.
train_Y = transfomed_label.reshape(-1,1)
```

```python
train_Y
```

```python
# Using "Logistic Regression" here....
import sklearn as sk  
from sklearn.linear_model import LogisticRegression  
import os

LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr').fit(train_X, train_Y.ravel())
LR.predict(train_X) 
print("Logistic Regression")
round(LR.score(train_X,train_Y), 4)
```

```python
# Using "Support Vector Machines" here....
from sklearn import svm  
SVM = svm.LinearSVC()
SVM.fit(train_X, train_Y)
SVM.predict(train_X)
print("Support Vector Machines")
round(SVM.score(train_X,train_Y), 4)
```

```python
# Using "Random Forest Classifier" here....
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)  
RF.fit(train_X, train_Y)  
RF.predict(train_X) 
print("Random Forests")
round(RF.score(train_X,train_Y), 4)
```

```python
# Using "MLP Classifier" here....
from sklearn.neural_network import MLPClassifier
NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)  
NN.fit(train_X, train_Y)  
NN.predict(train_X)
print("Neural Networks")
round(NN.score(train_X, train_Y), 4)
```

```python
#readinng the csv file of 'RAYMOND' stocks.
raymond = pd.read_csv (r'C:\Users\anuditakeshri\Desktop\CL\Small_Cap\RAYMOND.csv')
```

```python
#calculating :
# 1. day % change btw 'Close Price' & 'Open Price'
# 2. day % change btw 'High Price' & 'Low Price'
# 3. 5 day rolling mean of the day to day % change in 'Close Price'
# 4. 5 day rolling std of the day to day % change in 'Close Price'

raymond['%chg op_cl'] = ((raymond['Close Price'] - raymond ['Open Price'])/(raymond['Close Price']))*100
raymond['%chg lw_hg'] = ((raymond['High Price'] - raymond ['Low Price'])/(raymond['High Price']))*100
raymond['%chg 5dymean'] = raymond['Close Price'].pct_change().dropna().rolling(5).mean()
raymond['%chg 5dystd'] = raymond['Close Price'].pct_change().dropna().rolling(5).std()
```

```python
#checking for NaN values in all columns.
raymond.isna().sum()
```

```python
#dropping the NaN values.
raymond = raymond.dropna()
```

```python
#checking the no. of rows and columns.
raymond.shape
```

```python
#making a list of 'Close Price' of the dataframe.
arr = []
for value in raymond['Close Price'].items():
    arr.append(value[1])
```

```python
#checking the list
arr
```

```python
#checking the length of the list
len(arr)
```

```python
#resetting the index for operations ahead.
raymond = raymond.reset_index(drop=True)
```

```python
#making a list of 'Action' values by iterating.
val=[]
for i in range(0,488):
    if arr[i+1] > arr[i]:
        val.append(1)
    else:
        val.append(-1)
```

```python
#checking the length of the list
len(val)
```

```python
#saving the values of the list into a dataframe column named 'Action'.
raymond['Action'] = pd.DataFrame(val)
```

```python
#dropping all the NaN values.
raymond = raymond.dropna()
raymond
```

```python
#creating training input and output arrays.
train_X = raymond[['%chg op_cl','%chg lw_hg','%chg 5dymean','%chg 5dystd']]
train_Y = raymond[['Action']]
```

```python
# creating model....
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)  
RF.fit(train_X, train_Y)  
RF.predict(train_X) 
RF.score(train_X,train_Y) #accuracy score
```
